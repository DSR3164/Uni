### **Асимптотический анализ турнирного дерева (Tournament Tree):**

1. **Построение дерева:**  
   - Высота дерева: \(h = \lceil \log_2 n \rceil\).  
   - Число внутренних узлов: \(n - 1\).  
   - Сложность: \(O(n)\).  

2. **Обновление элемента:**  
   - Обновляется путь от листа к корню.  
   - Сложность: \(O(\log n)\).  

3. **Поиск победителя:**  
   - Победитель находится в корне.  
   - Сложность: \(O(1)\).  

4. **Удаление победителя:**  
   - Новый победитель поднимается за высоту дерева.  
   - Сложность: \(O(\log n)\).  

---

**Итоговые сложности:**  
| Операция               | Сложность       |  
|------------------------|-----------------|  
| Построение дерева      | \(O(n)\)        |  
| Обновление элемента    | \(O(\log n)\)   |  
| Поиск победителя       | \(O(1)\)        |  
| Удаление победителя    | \(O(\log n)\)   |  

**Применение:**  
- Слияние \(k\) отсортированных массивов.  
- Поиск минимального/максимального элемента.  
- Внешняя сортировка.
